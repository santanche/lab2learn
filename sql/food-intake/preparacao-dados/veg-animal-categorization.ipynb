# Importando Tabelas do FCID

```sql
DROP TABLE IF EXISTS Crop_Group_VegAni;

CREATE TABLE Crop_Group_VegAni (
  CGN VARCHAR(2),
  CGL VARCHAR(6),
  Crop_Group_Description VARCHAR(80),
  Veg_Animal CHAR(1),
  PRIMARY KEY (CGL)
) AS SELECT
  CGN, CGL, Crop_Group_Description, Veg_Animal
FROM CSVREAD('../../../data/food-intake/basics/FCID_Cropgroup_Description-veg.csv');

CREATE TABLE Intake_Subgroups (
  SeqN INTEGER NOT NULL,
  CG_SubGroup VARCHAR(6),
  Intake DECIMAL(13,7),
  PRIMARY KEY (SeqN, CG_SubGroup)
) AS SELECT
  SeqN, CG_SubGroup, Intake
FROM CSVREAD('../../../data/food-intake/grouped/intake-subgroups.csv');
```

```sql
SELECT * FROM Crop_Group_VegAni LIMIT 10;
```

```sql
SELECT * FROM Intake_Subgroups LIMIT 10;
```

```sql
DROP VIEW IF EXISTS Intake_VegAni;

CREATE VIEW Intake_VegAni AS
SELECT I.SEQN, V.Veg_Animal, SUM(I.Intake) Intake
FROM Intake_Subgroups I, Crop_Group_VegAni V
WHERE I.CG_SubGroup = V.CGL
GROUP BY I.SEQN, V.Veg_Animal
ORDER BY I.SEQN, V.Veg_Animal;
```

```sql
SELECT * FROM Intake_VegAni LIMIT 10;
```
